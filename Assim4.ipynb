```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([2, 4, 5, 3, 6, 5, 7])
y = np.array([35, 60, 20, 50, 50, 55, 60])
```

```python
df = pd.DataFrame({'Weight': x, 'Price': y})
df
```

```python
X = df[['Weight']] #2D
y = df['Price']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
M = model.coef_[0]
print(f"Slope (M): {M}")
```

```python
C = model.intercept_
print(f"Y-Intercept (C): {C}")
```

```python
predicted_price_for_6 = M * 6 + C
print(f"Predicted price for weight 6: {predicted_price_for_6}")
```

```python
df['Predicted_Price'] = model.predict(X)
df['Residual'] = df['Price'] - df['Predicted_Price']
print(f"Residuals: \n{df['Residual']}")
```

```python
MSE = np.mean(df['Residual'] ** 2)
print(f"Mean Squared Error (MSE): {MSE}")
```

```python
MAE = np.mean(np.abs(df['Residual']))
print(f"Mean Absolute Error (MAE): {MAE}")
```
